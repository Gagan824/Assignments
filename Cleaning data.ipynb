# importing mendatory libraries

```python
import pandas as pd
import re
from datetime import datetime
```

# reading data from sxcel file

```python
data  = pd.read_excel('messy.xlsx',sheet_name='test')
```

# fetching the columns from the data

```python
columns = data.columns
```

# here I am dropping off the column which is of no use

```python
data = data.drop(['Unnamed: 2'], axis =1)
```

# defining a function to correct the columns names 

```python
def column_correction(columns):
    new_columns = []
    for column in columns:
        string = re.sub(r'[^\w\s]','',column)
        string = string.lower()
        string = string.lstrip()
        string = string.replace(" ","_")
        new_columns.append(string)
        
    return new_columns
```

```python
columns = column_correction(columns)
```

# here I am setting up the columns name that we have corrected

```python
data = data.set_axis(columns, axis=1)
```

# converting date format to yyyy/mm/dd

```python
def convert_date(dates):
    new_date  = []
    for date in dates:
        if '-' in date or '/' in date:
            if '-' in date:
                date = date.replace("-","/")
                sdate = date.split(' ')
                new_date.append(datetime.strptime(sdate[0],"%Y/%m/%d"))
            else:
                new_date.append(datetime.strptime(date,"%y/%m/%d"))
        else:
            date = int(date)
            new_date.append(datetime.fromtimestamp(date))
        
    return new_date
```

```python
a =  convert_date(data['join_date'])
```

```python
data['join_date'] = a
```

```python
data
```

# Creating function to make name title case

```python
def making_title_name(names):
    new_names = []
    for name in names:
        new_names.append(name.title())
    return new_names
```

```python
names = making_title_name(data['fll_nam'])
```

```python
data['fll_nam'] = names
```

```python
data
```

# creating a function to create emails

```python
def create_email(data):
    emails = []
    for ind in data.index:
        name = data['fll_nam'][ind]
        sname = name.lower().split(' ')
        email = '{}.{}.{}@yourcompany.com'.format(sname[1],sname[0],data['cust_id'][ind])
        emails.append(email)
    return emails
```

```python
emails = create_email(data)
```

```python
data['email'] = emails
```

```python
data
```

# creating a function to correct the mobile numbers

```python
def correct_mobile_numbers(numbers):
    c_numbers = []
    for number in numbers:
        n = str(number)
        if '84' in n:
            c_numbers.append(int(n))
        else:
            c_n = '84'+n
            c_numbers.append(int(c_n))
            
    return c_numbers
```

```python
numbers = correct_mobile_numbers(data['mobiles'])
```

```python
data['mobiles'] = numbers
```
